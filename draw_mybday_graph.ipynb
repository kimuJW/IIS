```python
import csv

f= open('incheon_weather.csv', 'r')
data = csv.reader(f)
header = next(data)

low = []
average = []
high = []
date = []

for row in data:
    if len(row[0].strip().split('-')) > 1:
        year = row[0].strip().split('-')[0]
        month = row[0].strip().split('-')[1]
        day = row[0].strip().split('-')[2]
        if int(year) >= 1999:
              if month == '11' and day == '09':
                date.append(str(year) + '/' + str(month) + '/' + str(day))
                if row[-1] != '' and row[-2] != '' and row[-3] != '':
                    high.append(float(row[-1]))
                    low.append(float(row[-2]))
                    average.append(float(row[-3]))
print(date)
```

```python
plt.rc('font', family = 'Malgun Gothic')
plt.title("  ")

plt.plot(low, color='blue', label = '')
plt.plot(average, color='green', linestyle='--', marker='v', label = '')
plt.plot(high, color='red', label = '')
plt.xticks(range(len(date)), date, fontsize = 7, rotation=45)

plt.legend(loc=0)
plt.show()
```
