```python
import csv

f= open('incheon_weather.csv', 'r')
data = csv.reader(f)
header = next(data)

low = []
average = []
high = []
date = []

for row in data:
    if len(row[0].strip().split('-')) > 1:
        year = row[0].strip().split('-')[0]
        month = row[0].strip().split('-')[1]
        day = row[0].strip().split('-')[2]
        if int(year) >= 1999:
              if month == '11' and day == '09':
                date.append(str(year) + '/' + str(month) + '/' + str(day))
                if row[-1] != '' and row[-2] != '' and row[-3] != '':
                    high.append(float(row[-1]))
                    low.append(float(row[-2]))
                    average.append(float(row[-3]))
print(date)
```

평균온도를 저장할 새로운 리스트 average와 년도,월,일을 저장할 리스트 date를 새로 생성하여 
average에는 평균온도를 date에는 해당 년,월,일을 추가하도록 하였습니다.

```python
plt.rc('font', family = 'Malgun Gothic')
plt.title("내 생일부터 오늘까지")

plt.plot(low, color='blue', label = '최저온도')
plt.plot(average, color='green', linestyle='--', marker='v', label = '평균온도')
plt.plot(high, color='red', label = '최고온도')
plt.xticks(range(len(date)), date, fontsize = 7, rotation=45)

plt.legend(loc=0)
plt.show()
```

# 나의 생년월일부터 지금까지의 생일날의 기온을 그래프로 나타내었습니다.
 구분을 위해 최저온도는 파란색 평균온도는 초록색 최고온도는 빨간색으로 각각의 온도에 대한 상징적인 색을 사용,
선 모양과 마커는 솔직히 그대로 두는 편이 보기에 훨씬 좋은 것 같아서 평균온도만 연습해볼겸 
linestyle='--', marker='v'로 설정해보았습니다.
x축은 xticks를 사용하여 년도로 설정했는데 그냥 넣어버리면 문자열이 길어서 글자들이 겹치는 문제가 있었습니다. 따라서 폰트 사이즈를 적절하게 조정하고 rotation을 통해 문자열의 각도를 돌려서 겹치지 않도록 했습니다.
plt.xticks(range(len(date)), date, fontsize = 7, rotation=45) 해당 코드입니다.
